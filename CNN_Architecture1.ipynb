Q1.   What is the role of filters and feature maps in Convolutional Neural
Network (CNN)?

  Answer: In a Convolutional Neural Network (CNN), filters (also called kernels) and feature maps are the key components used to extract useful information from images.

Filters (Kernels): A filter is a small matrix (e.g., 3×3 or 5×5) that slides over the input image. It detects specific features such as edges, corners, textures, shapes, and patterns. During training, the values of these filters are automatically learned so that the CNN can recognize important visual features.
Feature Maps: When a filter is applied to an input image through the convolution operation, it produces an output called a feature map (activation map). A feature map highlights the locations where the corresponding feature is detected in the image. Different filters generate different feature maps, allowing the CNN to capture various characteristics of the image.

Role in CNN:

Filters extract important visual features from the input.
Feature maps store the detected features.
Multiple convolution layers learn increasingly complex features, from simple edges to complete objects.
These extracted features help the CNN accurately perform tasks such as image classification, object detection, and image recognition.


Q2. Explain the concepts of padding and stride in CNNs(Convolutional Neural
Network). How do they affect the output dimensions of feature maps?


Answer: Padding and Stride are two important parameters in a Convolutional Neural Network (CNN) that control how the convolution operation is performed and determine the size of the output feature maps.

1. Padding:
Padding is the process of adding extra rows and columns (usually zeros) around the border of the input image before applying the convolution filter. It helps preserve information at the image edges and prevents the output feature map from shrinking too quickly.

Valid Padding: No padding is added, so the output size becomes smaller.
Same Padding: Zero-padding is added so that the output feature map has approximately the same height and width as the input.

2. Stride:
Stride is the number of pixels by which the filter moves across the input image during convolution.

Stride = 1: The filter moves one pixel at a time, producing a larger and more detailed feature map.
Stride > 1: The filter skips pixels, resulting in a smaller feature map and reduced computation.

Effect on Output Dimensions:

Increasing padding increases or preserves the output size by adding borders to the input.
Increasing stride decreases the output size because the filter moves in larger steps.
Therefore, padding and stride directly affect the height and width of the output feature maps, as well as the computational cost and the amount of information retained.


Q3.  Define receptive field in the context of CNNs. Why is it important for deep
architectures?

Answer: The receptive field in a Convolutional Neural Network (CNN) is the region of the input image that influences the output of a particular neuron in a feature map. In other words, it represents the portion of the input image that a neuron "sees" to extract features.

In the early convolutional layers, the receptive field is small and captures simple features such as edges, corners, and textures. As the network becomes deeper, the receptive field increases, allowing neurons to capture larger and more complex patterns such as object parts and entire objects.

Importance for Deep Architectures:

A larger receptive field helps the network understand the global context of an image.
It enables the CNN to learn both local and high-level features effectively.
Deep architectures achieve better accuracy because deeper layers combine information from larger regions of the image.
This is especially important for tasks such as image classification, object detection, and image segmentation, where understanding the complete object is necessary.

Q4. Discuss how filter size and stride influence the number of parameters in a
CNN.

Answer: The filter size and stride are important hyperparameters in a CNN that affect its performance and computational efficiency.

Filter Size:

The filter size determines how much of the input image is processed at one time.
Larger filters (e.g., 5×5, 7×7) contain more weights, so they increase the number of trainable parameters and computational cost.
Smaller filters (e.g., 3×3) use fewer parameters, reduce computation, and are widely preferred in modern CNNs because multiple small filters can learn complex features efficiently.

Stride:

Stride determines how many pixels the filter moves during convolution.
A larger stride reduces the size of the output feature map and lowers computational cost.
However, stride does not change the number of trainable parameters, because the filter weights remain the same. It only changes the number of convolution operations and the output dimensions

Summary:

Larger filter size → More parameters and higher computation.
Smaller filter size → Fewer parameters and efficient feature extraction.
Larger stride → Smaller output feature maps and lower computation, but the number of trainable parameters remains unchanged.

Q6: Using keras, build and train a simple CNN model on the MNIST dataset
from scratch. Include code for module creation, compilation, training, and evaluation.

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocessing
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# Build CNN model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train, epochs=5, batch_size=64)

# Evaluate model
loss, accuracy = model.evaluate(x_test, y_test)

print("Test Accuracy:", accuracy)
```

Q7.  Load and preprocess the CIFAR-10 dataset using Keras, and create a

Q8.   Using PyTorch, write a script to define and train a CNN on the MNIST
dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Load MNIST dataset
transform = transforms.ToTensor()

train_dataset = datasets.MNIST(root='./data', train=True,
                               download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False,
                              download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Define CNN Model
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3)

        self.fc1 = nn.Linear(64 * 5 * 5, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64 * 5 * 5)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN()

# Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training Loop
epochs = 5

for epoch in range(epochs):
    model.train()
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}")

# Evaluation
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```

Q9.  Given a custom image dataset stored in a local directory, write code using
Keras ImageDataGenerator to preprocess and train a CNN model.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Data Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_data = train_datagen.flow_from_directory(
    "dataset/",
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_data = train_datagen.flow_from_directory(
    "dataset/",
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(train_data.num_classes, activation='softmax')
])

# Compile Model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train Model
model.fit(train_data,
          validation_data=val_data,
          epochs=10)

# Evaluate
loss, accuracy = model.evaluate(val_data)
print("Validation Accuracy:", accuracy)
```
